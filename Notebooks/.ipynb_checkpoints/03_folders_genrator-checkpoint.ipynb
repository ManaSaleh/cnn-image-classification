### **Step 2: Split the Data Using `ImageDataGenerator`**  

We will use TensorFlow's `ImageDataGenerator` to split the dataset into training, validation, and test sets.

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
# Initialize ImageDataGenerator for splitting data
data_gen = ImageDataGenerator(
    rescale=1.0/255,  # Normalize pixel values
    validation_split=0.2  # 80% training, 20% validation
)
```

```python
# Load and split the dataset
train_generator = data_gen.flow_from_directory(
    data_dir, 
    target_size=(150, 150),  # Resize images
    batch_size=32, 
    class_mode='binary',  # Binary classification: cats vs. dogs
    subset='training'  # Use training set
)
```

```python
validation_generator = data_gen.flow_from_directory(
    data_dir, 
    target_size=(150, 150), 
    batch_size=32, 
    class_mode='binary', 
    subset='validation'  # Use validation set
)
```

1. **Data Normalization:**  
   - `rescale=1.0/255` scales pixel values to `[0, 1]`.  

2. **Data Splitting:**  
   - `validation_split=0.2`: Reserves 20% of the data for validation.  
   - `subset='training'`: Selects the training portion.  
   - `subset='validation'`: Selects the validation portion.  

3. **Target Size:**  
   - `target_size=(150, 150)` resizes images to the expected model input size.  

4. **Batch Size and Class Mode:**  
   - `batch_size=32`: Processes 32 images at a time.  
   - `class_mode='binary'`: Labels as 0 (cat) or 1 (dog).

### **Step 3: Apply Data Augmentation Using `ImageDataGenerator`**  

We will apply various data augmentation techniques like rotation, flipping, zooming, and shifting using `ImageDataGenerator`. This helps increase dataset diversity, reducing overfitting.

```python
# Initialize ImageDataGenerator with data augmentation
train_augmented_gen = ImageDataGenerator(
    rescale=1.0/255,  # Normalize pixel values
    rotation_range=40,  # Random rotation
    width_shift_range=0.2,  # Horizontal shift
    height_shift_range=0.2,  # Vertical shift
    shear_range=0.2,  # Shear transformation
    zoom_range=0.2,  # Zoom in/out
    horizontal_flip=True,  # Flip images horizontally
    fill_mode='nearest'  # Fill missing pixels
)
```

```python
# Load augmented training data
train_augmented_generator = train_augmented_gen.flow_from_directory(
    data_dir, 
    target_size=(150, 150), 
    batch_size=32, 
    class_mode='binary', 
    subset='training'
)
```

```python
# Visualize augmented images
import matplotlib.pyplot as plt
```

```python
# Get a batch of augmented images
augmented_images, augmented_labels = next(train_augmented_generator)
```

```python
# Plot some augmented images
plt.figure(figsize=(12, 8))
for i in range(9):
    plt.subplot(3, 3, i+1)
    plt.imshow(augmented_images[i])
    plt.title(f"Label: {int(augmented_labels[i])}")
    plt.axis('off')

plt.show()
```

1. **Data Augmentation Settings:**  
   - `rotation_range=40`: Rotates images randomly by up to 40 degrees.  
   - `width_shift_range=0.2` & `height_shift_range=0.2`: Shifts images by up to 20%.  
   - `shear_range=0.2`: Applies shear transformation.  
   - `zoom_range=0.2`: Zooms in or out by up to 20%.  
   - `horizontal_flip=True`: Flips images horizontally.  
   - `fill_mode='nearest'`: Fills empty pixels created by transformations.

2. **Loading Augmented Data:**  
   - `flow_from_directory()` loads and applies augmentations on the fly.  

3. **Visualization:**  
   - `next(train_augmented_generator)` retrieves a batch of augmented images.  
   - We display 9 samples using `plt.imshow()`.
